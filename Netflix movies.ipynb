Import Statistical Libraries

```python
import statistics as stat
from statsmodels.stats.weightstats import ztest
```

Measures of Central Tendency and dispersion

Finding Statistical values for netflix movie Scores

```python
stat.mean(netflix['Score'])
```

```python
stat.variance(netflix['Score'])
```

```python
stat.median(netflix['Score'])
```

```python
stat.mode(netflix['Score'])
```

```python
stat.stdev(netflix['Score'])
```

Z test between Rank and Score 

```python
z_stat,p_val=ztest(netflix['Rank'],netflix['Score'],value=0)
p_val
```

```python
if p_val<0.05:
    print('Reject Null Hypothesis')
else:
    print('Accept Null Hypothesis')
  
```
